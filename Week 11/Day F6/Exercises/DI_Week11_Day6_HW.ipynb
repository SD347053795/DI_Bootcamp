```python
import numpy as np
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]
data_array = np.array(data)
mean = np.mean(data_array)
median = np.median(data_array)
variance = np.var(data_array, ddof=1)
std_deviation = np.std(data_array, ddof=1)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_deviation = 10
x = np.linspace(mean - 4*std_deviation, mean + 4*std_deviation, 1000)
pdf = norm.pdf(x, mean, std_deviation)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Normal Distribution (mean={mean}, std={std_deviation})', color='blue')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import kurtosis, skew
import matplotlib.pyplot as plt


data = np.random.normal(0, 1, 1000)
kurt = kurtosis(data)
skw = skew(data)
print(f"Kurtosis: {kurt}")
print(f"Skewness: {skw}")
```
