```python
import numpy as np
import pandas as pd

np.random.seed(0)
cities = ['Jerusalem', 'Tel Aviv', 'Hiafa', 'Ashdod', 'Ariel',
          'Karmiel', 'Beni Brak', 'Acco', 'Netanya', 'Hevron']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperatures = np.random.uniform(-5, 35, size=(10, 12))
df_temperatures = pd.DataFrame(temperatures, index=cities, columns=months)

print("Synthetic Temperature Data:")
print(df_temperatures)
```

```python
annual_avg_temp = df_temperatures.mean(axis=1)
hottest_city = annual_avg_temp.idxmax()
coldest_city = annual_avg_temp.idxmin()

print("\nAnnual Average Temperatures:")
print(annual_avg_temp)

print(f"\nCity with Highest Average Temperature: {hottest_city}")
print(f"City with Lowest Average Temperature: {coldest_city}")
```

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go

fig = go.Figure()

for city in df_temperatures.index:
    fig.add_trace(go.Scatter(
        x=months,
        y=df_temperatures.loc[city],
        mode='lines',
        name=city,
        hoverinfo='y+name'
    ))
fig.update_layout(
    title='Annual Average Temperatures',
    xaxis_title='Month',
    yaxis_title='Temperature (°C)',
    xaxis=dict(tickangle=45),
    legend=dict(orientation='h'),
    margin=dict(l=0, r=0, t=60, b=0),
    hovermode='x unified'
)
fig.show()



plt.figure(figsize=(10, 6))
plt.bar(df_temperatures.index, annual_avg_temp, color='skyblue')
plt.title('Annual Average Temperatures')
plt.xlabel('City')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
