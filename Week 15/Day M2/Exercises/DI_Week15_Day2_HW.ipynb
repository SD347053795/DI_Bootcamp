```python
import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

nlp = spacy.load('en_core_web_sm')

def preprocess_text(data):
    lemmatizer = WordNetLemmatizer()

    def process_review(review):
        review = review.lower()
        tokens = word_tokenize(review)
        tokens = [word for word in tokens if word.isalnum()]
        tokens = [lemmatizer.lemmatize(token) for token in tokens]
        processed_review = ' '.join(tokens)
        return processed_review

    preprocessed_reviews = [process_review(review) for review in data['Review']]
    preprocessed_data = {'Review': preprocessed_reviews}

    return data, preprocessed_data

def perform_ner(text):
    doc = nlp(text)
    entities = [(entity.text, entity.label_) for entity in doc.ents]
    return entities

def perform_pos_tagging(text):
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    return pos_tags

data = {
    'Review': [
        'At McDonald\'s the food was ok and the service was bad.',
        'I would not recommend this Japanese restaurant to anyone.',
        'I loved this restaurant when I traveled to Thailand last summer.',
        'The menu of Loving has a wide variety of options.',
        'The staff was friendly and helpful at Google\'s employees restaurant.',
        'The ambiance at Bella Italia is amazing, and the pasta dishes are delicious.',
        'I had a terrible experience at Pizza Hut. The pizza was burnt, and the service was slow.',
        'The sushi at Sushi Express is always fresh and flavorful.',
        'The steakhouse on Main Street has a cozy atmosphere and excellent steaks.',
        'The dessert selection at Sweet Treats is to die for!'
    ]
}

raw_data, preprocessed_data = preprocess_text(data)

print("Raw Data:")
for review in raw_data['Review']:
    print(review)

print("\nPreprocessed Data:")
for review in preprocessed_data['Review']:
    print(review)

sample_text = "At McDonald's the food was ok and the service was bad."
entities = perform_ner(sample_text)
pos_tags = perform_pos_tagging(sample_text)

print("\nNamed Entities:")
for entity, label in entities:
    print(f"Entity: {entity}, Label: {label}")

print("\nPart-of-Speech Tags:")
for token, pos in pos_tags:
    print(f"Token: {token}, POS: {pos}")
```

```python
def analyze_data(raw_data, preprocessed_data):
    for i, (raw_review, preprocessed_review) in enumerate(zip(raw_data['Review'], preprocessed_data['Review'])):
        print(f"Review {i+1}:")
        print(f"Raw Review: {raw_review}")

        raw_entities = perform_ner(raw_review)
        raw_pos_tags = perform_pos_tagging(raw_review)

        print("\nRaw Data - Named Entities:")
        for entity, label in raw_entities:
            print(f"Entity: {entity}, Label: {label}")

        print("\nRaw Data - POS Tags:")
        for token, pos in raw_pos_tags:
            print(f"Token: {token}, POS: {pos}")

        print(f"\nPreprocessed Review: {preprocessed_review}")

        preprocessed_entities = perform_ner(preprocessed_review)
        preprocessed_pos_tags = perform_pos_tagging(preprocessed_review)

        print("\nPreprocessed Data - Named Entities:")
        for entity, label in preprocessed_entities:
            print(f"Entity: {entity}, Label: {label}")

        print("\nPreprocessed Data - POS Tags:")
        for token, pos in preprocessed_pos_tags:
            print(f"Token: {token}, POS: {pos}")

        print("\n" + "-"*50 + "\n")

analyze_data(raw_data, preprocessed_data)
```

```python
# Import necessary libraries
from gensim.models import Word2Vec

# Step 1: Preprocessed and tokenized dataset
sentences = [
    ["this", "is", "the", "first", "sentence"],
    ["this", "is", "the", "second", "sentence"],
    ["yet", "another", "sentence"],
    ["one", "more", "sentence"],
    ["and", "the", "final", "sentence"]
]

# Step 2: Train the Word2Vec model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Step 3: Print the dimensions of the Word2Vec object
vector_dimensions = model.wv.vector_size
vocabulary_size = len(model.wv)

print(f"Vector Dimensions: {vector_dimensions}")
print(f"Vocabulary Size: {vocabulary_size}")

# Step 4: Analysis of the output
# - Vector Dimensions: The number of dimensions each word vector has (100 in this case).
# - Vocabulary Size: The number of unique words in the dataset (11 in this case).

# The vector dimensions indicate the length of the vector that represents each word.
# A larger number of dimensions allows for capturing more detailed semantic relationships,
# while a smaller number may lead to more general representations. The vocabulary size
# reflects the number of unique words that the model has learned from the provided dataset.
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def plot_word_embeddings(model):
    words = list(model.wv.index_to_key)
    vectors = model.wv[words]

    pca = PCA(n_components=2)
    reduced_vectors = pca.fit_transform(vectors)

    plt.figure(figsize=(12, 8))
    plt.scatter(reduced_vectors[:, 0], reduced_vectors[:, 1], marker='o')

    for i, word in enumerate(words):
        plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]), fontsize=12, ha='right')

    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('Word Embeddings Visualization')
    plt.grid(True)
    plt.show()

from gensim.models import Word2Vec

model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
plot_word_embeddings(model)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

sentences = [
    preprocess("This is the first sentence."),
    preprocess("This is the second sentence."),
    preprocess("Yet another sentence."),
    preprocess("One more sentence."),
    preprocess("And the final sentence.")
]

model = Word2Vec(sentences, vector_size=150, window=10, min_count=2, sg=1, workers=4)

def plot_word_embeddings(model):
    words = list(model.wv.index_to_key)
    vectors = model.wv[words]

    if len(words) > 1:
        try:
            tsne = TSNE(n_components=2, perplexity=min(5, len(words) - 1), random_state=0)
            reduced_vectors = tsne.fit_transform(vectors)
        except ValueError as e:
            print(f"t-SNE error: {e}. Using PCA instead.")
            pca = PCA(n_components=2)
            reduced_vectors = pca.fit_transform(vectors)
    else:
        print("Not enough words to perform dimensionality reduction.")
        return

    plt.figure(figsize=(12, 8))
    plt.scatter(reduced_vectors[:, 0], reduced_vectors[:, 1], marker='o')

    for i, word in enumerate(words):
        plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]), fontsize=12, ha='right')

    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.title('Word Embeddings Visualization')
    plt.grid(True)
    plt.show()

plot_word_embeddings(model)
```
