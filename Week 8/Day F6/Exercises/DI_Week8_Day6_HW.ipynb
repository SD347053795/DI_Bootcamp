1. Explain why data visualization is important in data analysis.

1A. Data visualization is crucial in data analysis because it allows analysts to present complex datasets in a clear and understandable manner. By visualizing data through charts, graphs, and other graphical representations, patterns, trends, and relationships within the data can be quickly identified. This not only aids in making data-driven decisions but also facilitates effective communication of findings to stakeholders, enhancing overall comprehension and insight extraction from the data.

2. Describe the purpose of a line graph in data visualization.

2A. A line graph serves the purpose of illustrating trends and relationships between continuous variables over a period of time or across different categories. It is particularly useful for showing changes in data over time and highlighting patterns such as growth, decline, or fluctuations. By plotting data points and connecting them with lines, a line graph provides a visual depiction of how a variable changes in relation to another, making it easier to interpret trends and make comparisons. Thus, line graphs are valuable tools in data visualization for revealing temporal patterns and trends in datasets.

```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [72, 74, 76, 80, 82, 78, 75]
plt.figure(figsize=(8, 5))  # Optional: set figure size
plt.plot(days, temperatures, marker='o', linestyle='-', color='b', label='Temperature (°F)')
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.title('Temperature Variation Over a Week')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [5000, 5500, 6200, 7000, 7500]
plt.figure(figsize=(8, 5))
plt.bar(months, sales, color='skyblue')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales for Retail Store')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'sales_data.csv'
df = pd.read_csv(file_path)

print(df.head())
total_quantity_sold = df['quantity'].sum()
print(f"Total quantity of products sold: {total_quantity_sold}")

category_revenue = df.groupby('category')['revenue'].sum()
highest_revenue_category = category_revenue.idxmax()
highest_revenue = category_revenue.max()
print(f"Category with highest revenue: {highest_revenue_category}")
print(f"Revenue generated by the highest revenue category: ${highest_revenue:.2f}")

total_sales = df.shape[0]
total_revenue = df['revenue'].sum()
average_revenue_per_sale = total_revenue / total_sales
print(f"Average revenue per sale: ${average_revenue_per_sale:.2f}")
df['date'] = pd.to_datetime(df['date'])
df['Quarter'] = df['date'].dt.quarter
quarterly_revenue = df.groupby('Quarter')['revenue'].sum()
plt.figure(figsize=(8, 5))
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
plt.bar(quarters, quarterly_revenue, color='skyblue')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue Generated in Each Quarter')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Line Plot
x = np.linspace(-10, 10, 100)
y = x ** 2
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='blue')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Bar Chart
products = ['A', 'B', 'C', 'D']
sales_values = [15, 30, 45, 20]
plt.figure(figsize=(8, 5))
plt.bar(products, sales_values, color='green')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Weekly Sales of Products')
plt.grid(True)
plt.show()

# Pie Chart
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['Green', 'Yellow', 'Red', 'Brown']
plt.figure(figsize=(8, 5))
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Favorite Fruits')
plt.axis('equal')
plt.legend(loc='best')
plt.show()
```
