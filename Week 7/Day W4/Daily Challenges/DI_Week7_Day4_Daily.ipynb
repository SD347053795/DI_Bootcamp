```python
from google.colab import files
uploaded = files.upload()

from PIL import Image, ImageOps
import matplotlib.pyplot as plt

def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()

image_path = list(uploaded.keys())[0]
original_image = Image.open(image_path)

rotated_image = original_image.rotate(90, expand=True)

flipped_horizontally_image = ImageOps.mirror(original_image)
flipped_vertically_image = ImageOps.flip(original_image)

width, height = original_image.size
zoomed_image = original_image.resize((int(width * 1.2), int(height * 1.2)))

show_images(original_image, rotated_image, 'Rotated Image')

show_images(original_image, flipped_horizontally_image, 'Horizontally Flipped Image')

show_images(original_image, flipped_vertically_image, 'Vertically Flipped Image')

show_images(original_image, zoomed_image, 'Zoomed Image')
```

```python
import os
import numpy as np
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

def load_and_visualize_images(image_dir):
    image_files = os.listdir(image_dir)
    images = [Image.open(os.path.join(image_dir, img_file)) for img_file in image_files]

    original_image = images[np.random.randint(len(images))]

    rotated_image = original_image.rotate(90)

    flipped_image = ImageOps.flip(ImageOps.mirror(original_image))

    zoomed_image = original_image.resize((int(original_image.width * 1.2), int(original_image.height * 1.2)))

    show_images(original_image, rotated_image, "Rotated Image")
    show_images(original_image, flipped_image, "Flipped Image")
    show_images(original_image, zoomed_image, "Zoomed Image")

def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()

if __name__ == "__main__":
    image_dir = "C:/Users/Shmuel/Desktop/DI_Bootcamp/Week_7/Day_W4/Flower_Folder/flower_images/flower_images"
    load_and_visualize_images(image_dir)
```
