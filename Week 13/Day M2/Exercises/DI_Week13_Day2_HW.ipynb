```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

chrome_options = Options()
chrome_options.add_argument("--headless")
service = Service('/usr/bin/chromedriver')
driver = webdriver.Chrome(service=service, options=chrome_options)

def scrape_rotten_tomatoes():
    try:
        driver.get('https://www.rottentomatoes.com/top/bestofrt/')
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '.table'))
        )
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')
        movies = soup.select('.table tr')
        for movie in movies:
            title = movie.select_one('.title')
            score = movie.select_one('.tMeterScore')
            release_date = movie.select_one('.release_date')

            if title and score and release_date:
                print(f"Title: {title.get_text(strip=True)}")
                print(f"Score: {score.get_text(strip=True)}")
                print(f"Release Date: {release_date.get_text(strip=True)}")
                print('---')

    finally:
        driver.quit()

if __name__ == "__main__":
    scrape_rotten_tomatoes()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from collections import defaultdict
from datetime import datetime
import re

chrome_options = Options()
chrome_options.add_argument("--headless")
service = Service('path/to/chromedriver')
driver = webdriver.Chrome(service=service, options=chrome_options)

def scrape_news():
    try:
        driver.get('https://www.bbc.com/news/technology')
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '.gs-c-promo'))
        )
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')
        articles = soup.select('.gs-c-promo')
        categorized_articles = defaultdict(list)

        for article in articles:
            title_element = article.select_one('.gs-c-promo-heading__title')
            date_element = article.select_one('.gs-o-bullet')
            if title_element and date_element:
                title = title_element.get_text(strip=True)
                date_text = date_element.get_text(strip=True)
                try:
                    date_match = re.search(r'\d{1,2} \w+ \d{4}', date_text)
                    if date_match:
                        date_obj = datetime.strptime(date_match.group(0), '%d %B %Y')
                        month_name = date_obj.strftime('%B')
                        categorized_articles[month_name].append(title)
                except ValueError:
                     print(f"Date parsing error for: {date_text}")
        for month, titles in categorized_articles.items():
            print(f"\nArticles from {month}:")
            for title in titles:
                print(f"- {title}")

    finally:
        driver.quit()

if __name__ == "__main__":
    scrape_news()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from collections import Counter
import numpy as np

chrome_options = Options()
chrome_options.add_argument("--headless")
service = Service('path/to/chromedriver')
driver = webdriver.Chrome(service=service, options=chrome_options)

def scrape_weather():
    try:
        driver.get('https://www.bbc.com/weather/293397')
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '.weather-forecast'))
        )
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')
        temperatures = []
        conditions = []
        humidities = []
        forecast_items = soup.select('.forecast-item')
        for item in forecast_items:
            temp_element = item.select_one('.temperature')
            condition_element = item.select_one('.condition')
            humidity_element = item.select_one('.humidity')

            if temp_element and condition_element and humidity_element:
                try:
                    temperature = float(temp_element.get_text(strip=True).replace('°', ''))
                    condition = condition_element.get_text(strip=True)
                    humidity = float(humidity_element.get_text(strip=True).replace('%', ''))

                    temperatures.append(temperature)
                    conditions.append(condition)
                    humidities.append(humidity)
                except ValueError:
                    print(f"Conversion error for temperature or humidity: {temp_element}, {humidity_element}")

        if temperatures:
            avg_temperature = np.mean(temperatures)
            most_common_condition = Counter(conditions).most_common(1)[0][0]
            print(f"Average Temperature: {avg_temperature:.2f}°C")
            print(f"Most Common Weather Condition: {most_common_condition}")
        else:
            print("No weather data found.")

    finally:
        driver.quit()

if __name__ == "__main__":
    scrape_weather()
```
