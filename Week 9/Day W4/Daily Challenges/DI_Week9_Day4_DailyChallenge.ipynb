```python
import pandas as pd

file_path = 'US Superstore data.xls'
df = pd.read_excel(file_path)

df.drop_duplicates(inplace=True)
print("DataFrame shape after removing duplicates:", df.shape)

df.dropna(inplace=True)
print("DataFrame shape after removing null cells:", df.shape)

cleaned_file_path = 'Cleaned_US_Superstore_data.csv'
df.to_csv(cleaned_file_path, index=False)

print("Cleaned data saved to:", cleaned_file_path)
```

```python
import pandas as pd
import plotly.express as px

file_path = 'Cleaned_US_Superstore_data.csv'
df = pd.read_csv(file_path)

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year

sales_trends = df.groupby('Year')['Sales'].sum().reset_index()
fig_line = px.line(sales_trends, x='Year', y='Sales', title='Sales Trends Over the Years')
fig_line.update_layout(xaxis=dict(dtick=1))

fig_line.show()

sales_distribution = df.groupby('Country')['Sales'].sum().reset_index()
fig_map = px.choropleth(sales_distribution,
                        locations='Country',
                        locationmode='country names',
                        color='Sales',
                        title='Sales Distribution by Country')
fig_map.show()
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, geom_point, theme_minimal, labs

file_path = 'Cleaned_US_Superstore_data.csv'
df = pd.read_csv(file_path)

df['Order Date'] = pd.to_datetime(df['Order Date'])
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10).reset_index()
top_products = top_products.sort_values(by='Sales', ascending=False)

bar_chart = (
    ggplot(top_products, aes(x='Product Name', y='Sales')) +
    geom_bar(stat='identity', fill='skyblue') +
    theme_minimal() +
    labs(title='Top 10 Products by Sales', x='Product Name', y='Sales')
)
print(bar_chart)

scatter_plot = (
    ggplot(df, aes(x='Discount', y='Profit')) +
    geom_point(alpha=0.5, color='blue') +
    theme_minimal() +
    labs(title='Relationship Between Profit and Discount', x='Discount', y='Profit')
)
print(scatter_plot)
```
