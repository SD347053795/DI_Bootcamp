```python
import pandas as pd
import numpy as np
import random
from plotnine import ggplot, aes, geom_point, labs, theme_minimal

np.random.seed(0)
days = pd.date_range('2023-01-01', periods=20, freq='D')
stock_prices = np.random.normal(100, 20, 20).round(2)

df = pd.DataFrame({'Date': days, 'Stock Price': stock_prices})

scatter_plot = ggplot(df, aes(x='Date', y='Stock Price')) + \
               geom_point() + \
               labs(x='Date', y='Stock Price', title='Stock Prices Over Time') + \
               theme_minimal()
print(scatter_plot)
```

```python
import pandas as pd
import random
from plotnine import ggplot, aes, geom_bar, labs, theme_minimal, theme, element_text

reit_stocks = [
    'Simon Property Group', 'Prologis', 'Equity Residential', 'Public Storage',
    'Welltower', 'Digital Realty Trust', 'Ventas', 'AvalonBay Communities',
    'Boston Properties', 'Realty Income', 'Extra Space Storage', 'Healthpeak Properties',
    'Essex Property Trust', 'SL Green Realty', 'Kimco Realty', 'Federal Realty Investment Trust',
    'UDR', 'Iron Mountain', 'Equinix', 'STORE Capital'
]

random.seed(42)

data = {
    'REIT Stock': random.choices(reit_stocks, k=30)
}

df = pd.DataFrame(data)

bar_plot = ggplot(df, aes(x='REIT Stock')) + \
           geom_bar() + \
           labs(x='REIT Stock', y='Frequency', title='REIT Stock Frequency') + \
           theme_minimal() + \
           theme(axis_text_x=element_text(angle=45, hjust=1))

print(bar_plot)
```

```python
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_line, labs, theme_minimal

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

df = pd.DataFrame({
    'x': x,
    'y1': y1,
    'y2': y2
})

line_plot = ggplot(df, aes(x='x')) + \
            geom_line(aes(y='y1'), color='blue', size=1.5, linetype='solid') + \
            geom_line(aes(y='y2'), color='red', size=1.5, linetype='dashed') + \
            labs(x='X-Value', y='Y-Value', title='Comparison of Two Functions') + \
            theme_minimal()

line_plot = line_plot + labs(color='Function', linetype='Function') + \
            labs(color='sin(x)' , linetype='cos(x)')

print(line_plot)
```

```python
import plotly.graph_objects as go

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]

fig = go.Figure()

fig.add_trace(go.Scatter(x=months, y=sales, mode='lines+markers', name='Sales'))

fig.update_layout(title='Monthly Sales Data',
                  xaxis_title='Month',
                  yaxis_title='Sales')

fig.show()
```

```python
import plotly.graph_objects as go

genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]
fig = go.Figure()
fig.add_trace(go.Bar(x=genres, y=books_sold, marker_color='royalblue', name='Books Sold'))
fig.update_layout(title='Books Sold by Genre',
                  xaxis_title='Genre',
                  yaxis_title='Number of Books Sold')

fig.show()
```

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})

fig = px.scatter(data, x='Age', y='Salary', size='Years of Experience',
                 title='Relationship between Age, Salary, and Experience',
                 labels={'Age': 'Age', 'Salary': 'Salary', 'Years of Experience': 'Years of Experience'},
                 size_max=15)

fig.show()
```
